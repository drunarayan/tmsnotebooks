# Calculate Diameter of an Atom!



### 1. Make a 1/400 Dilute solution of Oleic Acid


```python
amount_of_oleicacid_in_one_mili_liter_of_alchohol = 0.042
print(amount_of_oleicacid_in_one_mili_liter_of_alchohol)
```

### 2. Measure the number of drops in one mililiter of alchohol

```python
known_volume_of_alchohol = 1.05 # ml
no_of_drops_in_known_volume_of_alchohol = 25 # We need to find this experimentally!
print(no_of_drops_in_known_volume_of_alchohol, "drops")
```

### 3. Calculate the amount of Oleic acid in one drop of dilution solution

```python
oleic_acid_in_one_drop_of_solution_ml = 0.003139
print(oleic_acid_in_one_drop_of_solution_ml, "ml")
```

### 4. Measure the Mono Molecular Layer!

```python
diameter_mono_molecular_layer_cm = 27.5
print(diameter_mono_molecular_layer_cm, "cm")
```

### 5. Calculate the area of the Mono Molecular Layer Disk


```python
import math
Area_of_Mono_Molecular_Disk = math.pi * diameter_mono_molecular_layer_cm**2 / 4
print(Area_of_Mono_Molecular_Disk, "cm^2")
```

## 6. Calculate the Thickness of the Mono Molecular Disk.
## This is the Volume of the Oleic Acid in one droplet divided by the Area of the Mono Molecular Disk!

```python
oleic_acid_in_one_drop_of_solution_litres = oleic_acid_in_one_drop_of_solution_ml * 1e-3
thickness_of_mono_molecular_layer = oleic_acid_in_one_drop_of_solution_litres / Area_of_Mono_Molecular_Disk
print(thickness_of_mono_molecular_layer, "cm")
no_of_hydrophobic_hydro_carbon_atoms_in_a_oleic_acid_molecule = 18
print(f'{no_of_hydrophobic_hydro_carbon_atoms_in_a_oleic_acid_molecule} Carbon Atoms')
```

### 7. Calculate the Diameter of Carbon Atom !!


```python
diameter_of_a_single_carbon_atom = thickness_of_mono_molecular_layer /  no_of_hydrophobic_hydro_carbon_atoms_in_a_oleic_acid_molecule
print(f'{diameter_of_a_single_carbon_atom:0.3e} cm or {(diameter_of_a_single_carbon_atom/100):0.3e}')
```
