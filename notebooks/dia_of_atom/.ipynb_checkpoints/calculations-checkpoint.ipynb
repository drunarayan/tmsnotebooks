# Calculate Diameter of Sun!

```python
import math
```

```python
#Average Time Sun takes to move by 1 diameter
traverse_time = 131.825 #seconds
print(traverse_time)
```

```python
# No of seconds in a 24-hour day
sec_in_a_day = 24 * 60 * 60  #seconds
print(sec_in_a_day)
```

```python
# Angular Distance of Sun from Earth in Deg
angular_distance_deg = traverse_time * 360 / sec_in_a_day
print(angular_distance_deg, "Deg")
```

```python
# Angular Distance of Sun from Earth in Rad
angular_distance_rad = angular_distance_deg * math.pi / 180
print(angular_distance_rad, "Rad")
```

```python
# Distance of Sun from Earth
sun_distance_from_earth = 150e6 # KM
```

```python
# Diameter of Sun Calculation
# For very Small angles the subtended arc of circle
# is equal to side of angle * angle in radians

diameter_of_sun = angular_distance_rad * sun_distance_from_earth
print(diameter_of_sun, "KM")
```

# Calculate Diameter of an Atom!

### 1. Make a 1/400 Dilute solution of Oleic Acid

```python
amount_of_oleicacid_in_one_mili_liter_of_alchohol = 1/20*1/20
print(amount_of_oleicacid_in_one_mili_liter_of_alchohol)
```

### 2. Measure the number of drops in one mililiter of alchohol

```python
no_of_drops_in_one_mili_liter_of_alchohol = 42 # We need to find this experimentally!
print(no_of_drops_in_one_mili_liter_of_alchohol, "drops")
```

### 3. Calculate the amount of Oleic acid in one drop of dilution solution

```python
oleic_acid_in_one_drop_of_solution_ml = amount_of_oleicacid_in_one_mili_liter_of_alchohol / no_of_drops_in_one_mili_liter_of_alchohol  # We need to find this experimentally!
print(oleic_acid_in_one_drop_of_solution_ml, "ml")
oleic_acid_in_one_drop_of_solution = oleic_acid_in_one_drop_of_solution_ml *1e-6
print(oleic_acid_in_one_drop_of_solution, "m^3")
```

### 4. Measure the Mono Molecular Layer!

```python
diameter_mono_molecular_layer_in = 4.625 # IN We need to find this experimentally!
print(diameter_mono_molecular_layer_in, "in")
diameter_mono_molecular_layer_cm = diameter_mono_molecular_layer_in * 2.54 
print(diameter_mono_molecular_layer_cm, "cm")
diameter_mono_molecular_layer = diameter_mono_molecular_layer_cm * 1e-2
print(diameter_mono_molecular_layer, "m")
```

### 5. Calculate the area of the Mono Molecular Layer Disk


```python
Area_of_Mono_Molecular_Disk = math.pi * diameter_mono_molecular_layer**2 / 4
print(Area_of_Mono_Molecular_Disk, "m^2")
```

##6. Calculate the Thickness of the Mono Molecular Disk.
## This is the Volume of the Oleic Acid in one droplet divided by the Area of the Mono Molecular Disk!

```python
thickness_of_mono_molecular_layer = oleic_acid_in_one_drop_of_solution / Area_of_Mono_Molecular_Disk
print(thickness_of_mono_molecular_layer, "m")
no_of_carbon_atoms_in_a_oleic_acid_molecule = 17
print(no_of_carbon_atoms_in_a_oleic_acid_molecule, "Carbon Atoms")
```

### 7. Calculate the Diameter of Carbon Atom !!


```python
diameter_of_a_single_carbon_atom = thickness_of_mono_molecular_layer /  no_of_carbon_atoms_in_a_oleic_acid_molecule
print(diameter_of_a_single_carbon_atom, "m")
```
