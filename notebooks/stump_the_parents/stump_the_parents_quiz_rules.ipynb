## Rules of Stump the parents Quiz Rules: 
1.  All students ask parents questions. Students should know the answers themselves!
1. There is a Quiz Master elected by the students. Their decision will be FINAL - no arguments.  Sometimes, the Quiz Master may ask for a vote from the students in case of a dispute.
1. There is a Fair Score Keeper whose authority may not be questioned. 
1.  Parents/Adults will be divided into 2 teams.  Correct answers will get 5 points. Totally wrong answers will get -1 point.  No answers will get -2 points.
1.  Somewhat correct answers might get partial points - student asking the question will decide a number between 1 and 4.
1.  If a wrong or somewhat correct answer is given, the other team gets to answer the question also. If a correct or more accurate answer is given, the other team will get anywhere from 1 to 3 points based on the accuracy of the answer. There will be no negative points in this case for wrong or no answers.
1. The Quiz will end after 20 questions are asked or sooner if time is running out. The Quiz Master will decide.
1. The losing team's parents will get a surprise gift for all the students participting. The gifts have already been purchased.  The losing team just has to pay up.  The winning team parents may chhose to show mercy by also contributing.
