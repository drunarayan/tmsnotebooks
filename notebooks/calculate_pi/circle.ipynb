## Meaning of PI

PI is a universal constant (its value never changes) with a value of ```3.1415925``` accurate to 8 decimal places or 9 digits of precision. Only 5 digits of precision was needed to go to the Moon and back without any problems. A number hqving 8 digits of precision is a very accurate number indeed.  I shows up in a lot calculations wherever circles are present.  It also shows up sometimes when no circles are to be found.  

#### Now I want to tell you what it means.  

1. Draw a circle with a compass. Mark its center with a pencil dot. Now, carefully draw a straight line with a ruler through the center so that the line touches the circle at 2 points on each side of the center. This line called the ```dimeter```of the circle. The line from the center to the edge is called the ```radius```. The length of the diameter is twice the length of the radius.  You check this with your ruler. This much you probably already know.
1. The curved line that forms the circle itself has a special name, its called the ```circumference```. 
1. The ```circumference``` of any circle anywhere in the universe divided by its ```diameter``` gives the magical constant $\pi$. It is the most famous constant in all of mathematics!

#### Let's Construct a LARGE Circle and measure $\pi$

1. Get a flexible measuring tape or rope, measurre out a meter in length. I does not have to be accurate, let's say as tall as a 10-year old.  This will be the ```radius``` of the circle
1. Affix one end of the tape or rope to a point on the ground outside the house such that it will rotate about the point but not move from side to side.  You may try hammering a nail through a hole in the tape or rope to the ground. Someone with strong hands can put a forceful thumb on it.
1. The affix a piece of chalk on the end other end and draw a perfect circle on the ground.  It may take several attempts to get it just right with the curved line meeting where it started.
1. Then measure out the ```diameter``` of the circle by putting the tape or rope on one edge, through the center to the other edge.
1. Then use that ```diameter``` length of the tape or rope to carefully lay it out on the circle to match the graceful curve perfectly. Mark the starting point.
1. Do it repeatedly from end to end such that you returned to where you started. How many times did you have to lay out a ```diameter``` length?
1. You will notice that you will have to do it for slight more than 3 times to get to the beginning.
1. You have now estimated the value of $\pi$ to be slightly greater than 3.
1. As you already know the precise value is `3.1415925` accurate to 8 digits!
