## 🛠️ Part 1: Teacher Background & Geometry Derivations

### 1. The Real Definition of a Radian
A **radian** is a natural mathematical unit of angular measurement based entirely on the internal geometric layout of a circle. 

* **The Benchmark Definition:** 1 Radian is the exact internal angle formed at the center of a circle when its outer boundary **arc length** matches its **radius (r)** perfectly.
* **The Circular Path:** Because a full circle has a total circumference equal to 2 * Pi * r, a 360-degree rotation contains exactly 2 * Pi radians (approx 6.28 rad). A standard half-turn of 180 degrees equals exactly Pi radians (~3.14159 rad).

### 2. Pixel Scale Derivation via Telescope Hardware
Instead of guessing or estimating values, you can compute image limits directly using the physical attributes of your smart imaging gear:
* **Telescope Lens Focal Length (f):** 250 mm
* **Digital Camera Sensor Pixel Pitch (p):** 2.9 µm = 0.0029 mm

Taking the explicit spatial ratio of pixel width against focal focus length gives the exact rad metric value captured inside each individual pixel element container:

* **Pixel Footprint Value (Radians)** = Pixel Width / Focal Length = 0.0029 mm / 250 mm = **0.0000116 radians per pixel**

To upscale this raw measure into traditional astronomical arc-seconds, multiply the output radian factor by 206,265 (the arc-seconds inside one radian unit):
* **Derived Image Pixel Scale** = 0.0000116 rad * 206,265 = **2.3927 arc-seconds per pixel**

*(This calculation cleanly explains where our standard classroom shortcut constant of **2.4 arc-seconds per pixel** comes from!)*

### 3. The Small Angle Approximation Formula
The **Small Angle Formula** establishes the physical width line profile (D) of an astronomical body using its total distance away from our observation deck (d) and its absolute angular scale on the sky dome (θ) measured in radians:

* **True Physical Diameter (D)** = Distance to Object (d) * Angular Profile in Radians (θ)
