```python
import cv2
import numpy as np

# Configuration - update file path to your snapshot if needed
image_filename = "2024_solar_eclipse_peak.jpg" 
base_img = cv2.imread(image_filename)

# Automatic layout mockup generator if local file is missing
if base_img is None:
    base_img = np.zeros((900, 700, 3), dtype=np.uint8)
    cv2.circle(base_img, (350, 450), 300, (40, 40, 40), -1)
    cv2.circle(base_img, (350, 450), 295, (10, 10, 10), -1)
    cv2.circle(base_img, (350, 747), 8, (240, 100, 200), -1)

# Persistent configuration tracking arrays
angle = 0
clicks = []
img_working = base_img.copy()

def update_view():
    """Handles rotation array matrices and updates line positions instantly."""
    global img_working, angle, clicks
    h, w = base_img.shape[:2]
    
    if angle == 0:
        img_working = base_img.copy()
    else:
        center = (w // 2, h // 2)
        rot_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
        img_working = cv2.warpAffine(base_img, rot_matrix, (w, h))
        
    # Draw tracking boundaries (Yellow for Start Line, Red for End Line)
    for idx, x_pos in enumerate(clicks):
        color = (0, 255, 255) if idx == 0 else (0, 0, 255)
        cv2.line(img_working, (x_pos, 0), (x_pos, h), color, 2)
        
    cv2.imshow("Astronomy Lab Tool", img_working)

def click_callback(event, x, y, flags, param):
    global clicks
    if event == cv2.EVENT_LBUTTONDOWN:
        if len(clicks) < 2:
            clicks.append(x)
            print(f"[*] Placed line boundary marker {len(clicks)} at X = {x} px")
            update_view()
            
            if len(clicks) == 2:
                # Math calculations execute on the second click lock
                delta_px = abs(clicks[1] - clicks[0])
                print("\n" + "="*45)
                print("         LAB ANALYSIS LOGGED")
                print("="*45)
                print(f" -> Line A Boundary Position : {clicks[0]} px")
                print(f" -> Line B Boundary Position : {clicks[1]} px")
                print(f" --> COMPUTED SPATIAL PROFILE: {delta_px:.2f} PIXELS")
                print("="*45)
                print("[!] Finished! Press 'Q' or 'ESC' on your keyboard to close window.")

# Start background manager loop to prevent notebook thread lockouts
cv2.startWindowThread()
cv2.namedWindow("Astronomy Lab Tool", cv2.WINDOW_AUTOSIZE)
update_view()
cv2.setMouseCallback("Astronomy Lab Tool", click_callback)

print("="*50)
print("   ECLIPSE SCANNER LAB UTILITY (OPENCV ENGINE)")
print("="*50)
print(" Instructions:")
print("  * Left-Click twice on the frame to place vertical measurement bounds.")
print("  * Press 'R' on your keyboard to rotate image 10° clockwise.")
print("  * Press 'C' to clear drawn lines and restart tracker values.")
print("  * Press 'Q' or 'ESC' while focusing on window to shut it down.")
print("-"*50)

# Window event execution engine
while True:
    key = cv2.waitKey(1) & 0xFF
    if key == ord('r'):
        angle = (angle + 10) % 360
        clicks = []  # Auto-clear entries on rotation shift
        print(f"[i] Rotated layout tracking canvas to: {angle}°")
        update_view()
    elif key == ord('c'):
        clicks = []
        print("[i] Measurements cleared. System ready.")
        update_view()
    elif key in [ord('q'), 27]:  # 27 represents standard escape key code
        break

# Complete canvas flush loop to guarantee windows close out clean on Macs/PCs
cv2.destroyAllWindows()
for i in range(5):
    cv2.waitKey(1)
print("\n[✔] Active canvas screen session closed successfully.")
```
